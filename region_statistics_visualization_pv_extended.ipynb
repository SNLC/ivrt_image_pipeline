```python
csv_list = glob.glob("**/summarized_counts/summary.csv", recursive=True)
df_list = [pd.read_csv(f) for f in csv_list]
```

```python
summary_df = pd.concat(df_list)
```

```python
bool_mask = (summary_df['Target Cell Type'] == 'vip')

fig, ax = plt.subplots(3,3)
fig.set_size_inches(15, 12)
fig.suptitle("Summarized vip to pv connectivity", fontsize=20, y=1.08)

ax[0,0].scatter(summary_df.loc[bool_mask, 'starter'], summary_df.loc[bool_mask, 'presynaptic/starter'])
ax[0,0].set_title('Presynaptic cells/Starter vs. \n # of Starter cells in Region')
ax[0,0].set_ylabel('Presynaptic cells/Starter')
ax[0,0].set_xlabel('# of Starter cells in region')

ax[0,1].scatter(summary_df.loc[bool_mask, 'starter'], summary_df.loc[bool_mask, 'presynaptic vip/vip'])
ax[0,1].set_title('Presynaptic vip/vip vs. \n # of Starter cells in Region')
ax[0,1].set_ylabel('Presynaptic vip/vip')
ax[0,1].set_xlabel('# of Starter cells in region')

ax[0,2].scatter(summary_df.loc[bool_mask, 'starter'], summary_df.loc[bool_mask, 'presynaptic vip/presynaptic'])
ax[0,2].set_title('Presynaptic vip/presynaptic vs. \n # of Starter cells in Region')
ax[0,2].set_ylabel('Presynaptic vip/presynaptic')
ax[0,2].set_xlabel('# of Starter cells in region')

ax[1,0].scatter(summary_df.loc[bool_mask, 'presynaptic'], summary_df.loc[bool_mask, 'presynaptic/starter'])
ax[1,0].set_title('Presynaptic cells/Starter vs. \n # of presynaptic cells in Region')
ax[1,0].set_ylabel('Presynaptic cells/Starter')
ax[1,0].set_xlabel('# of presynaptic cells in Region')

ax[1,1].scatter(summary_df.loc[bool_mask, 'presynaptic'], summary_df.loc[bool_mask, 'presynaptic vip/vip'])
ax[1,1].set_title('Presynaptic vip/vip vs. \n # of presynaptic cells in Region')
ax[1,1].set_ylabel('Presynaptic vip/vip')
ax[1,1].set_xlabel('# of presynaptic cells in Region')

ax[1,2].scatter(summary_df.loc[bool_mask, 'presynaptic'], summary_df.loc[bool_mask, 'presynaptic vip/presynaptic'])
ax[1,2].set_title('Presynaptic vip/presynaptic vs. \n # of presynaptic cells in Region')
ax[1,2].set_ylabel('Presynaptic vip/presynaptic')
ax[1,2].set_xlabel('# of presynaptic cells in Region')

ax[2, 0].axis('off')

ax[2,1].scatter(summary_df.loc[bool_mask, 'presynaptic/starter'], summary_df.loc[bool_mask, 'presynaptic vip/vip'])
ax[2,1].set_title('Presynaptic vip/vip vs. \n Presynaptic cells/Starter')
ax[2,1].set_ylabel('Presynaptic vip/vip')
ax[2,1].set_xlabel('Presynaptic cells/Starter')

ax[2,2].scatter(summary_df.loc[bool_mask, 'presynaptic/starter'], summary_df.loc[bool_mask, 'presynaptic vip/presynaptic'])
ax[2,2].set_title('Presynaptic vip/presynaptic vs. \n Presynaptic cells/Starter')
ax[2,2].set_ylabel('Presynaptic vip/presynaptic')
ax[2,2].set_xlabel('Presynaptic cells/Starter')

plt.tight_layout()
```
