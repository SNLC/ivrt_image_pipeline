```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os
```

```python
cd /Volumes/My Book/rabies_tracing_images/som_cre_starter_cells
```

```python
csv_list = glob.glob("**/summarized_counts/summary.csv", recursive=True)
df_list = [pd.read_csv(f) for f in csv_list]
```

```python
summary_df = pd.concat(df_list)
```

```python
summary_df
```

```python
bool_mask = (summary_df['Target Cell Type'] == 'vip')

fig, ax = plt.subplots(1,3)
fig.set_size_inches(15, 4)
fig.suptitle("Summarized vip to sst connectivity", fontsize=20, y=1.08)

ax[0].scatter(summary_df.loc[bool_mask, 'starter'], summary_df.loc[bool_mask, 'presynaptic/starter'])
ax[0].set_title('Presynaptic cells/Starter vs. \n # of Starter cells in Region')
ax[0].set_ylabel('Presynaptic cells/Starter')
ax[0].set_xlabel('# of Starter cells in region')

ax[1].scatter(summary_df.loc[bool_mask, 'starter'], summary_df.loc[bool_mask, 'presynaptic vip/vip'])
ax[1].set_title('Presynaptic vip/vip vs. \n # of Starter cells in Region')
ax[1].set_ylabel('Presynaptic vip/vip')
ax[1].set_xlabel('# of Starter cells in region')

ax[2].scatter(summary_df.loc[bool_mask, 'starter'], summary_df.loc[bool_mask, 'presynaptic vip/presynaptic'])
ax[2].set_title('Presynaptic vip/presynaptic vs. \n # of Starter cells in Region')
ax[2].set_ylabel('Presynaptic vip/presynaptic')
ax[2].set_xlabel('# of Starter cells in region')

plt.tight_layout()
```
