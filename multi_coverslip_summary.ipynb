```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
path = ("/Volumes/My Book/rabies_tracing_images/"
        "vip_cre_starter_cells/ML_08_15_17/"
        "sst_presynaptic_647/summarized_counts")
file_name = "summary.csv"
file_path = os.path.join(path, file_name)
```

```python
summary_df = pd.read_csv(file_path)
target_cell_type = summary_df['Target Cell Type'][0]
```

```python
plt.scatter(summary_df['starter'], summary_df['presynaptic/starter'])
plt.title('Presynaptic cells/Starter vs. # of Starter cells in Region for ' + target_cell_type + ' Labeling')
plt.ylabel('Presynaptic cells/Starter')
plt.xlabel('# of Starter cells in region')
```

```python
for coverslip_id, df in summary_df.groupby('Coverslip ID'):
    plt.scatter(df['starter'], df['presynaptic ' + target_cell_type + '/presynaptic'])
    plt.title('Presynaptic ' + target_cell_type + '/presynaptic ' + 'vs. # of starter cells in Region')
    plt.ylabel('Presynaptic ' + target_cell_type + '/presynaptic')
    plt.xlabel('# of Starter cells in region')
    plt.ylim(0,1)
```
